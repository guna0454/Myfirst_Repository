```python
#global and local

#global variables example
#x and y variables globally defined but it taking as arguments to the method 

x = 5
y = 10
def sum():
   sum = x + y
   return sum
print(sum())

#local variables example
#locally we are passing the variable values to the method 
def sum(x,y):
   sum = x + y
   return sum
print(sum(5, 10))
```

```python
#string methods and how it works 

#1.capitalize()
#The capitalize() method returns a string where the first character is upper case, and the rest is lower case.

txt = "gunadracks"

x = txt.capitalize()

print (x)

txt = "python is FUN!"

x = txt.capitalize()

print (x)

txt = "36 is my age."

x = txt.capitalize()

print (x)

print('---------------------------------------------------------------------------------------')

#2.casefold()
#The casefold() method returns a string where all the characters are lower case.
#lowercase() and casefold() both are same only but casefold() is stronger and more aggressive and meaningful.
txt = "Hello, And Welcome To My World!"

x = txt.casefold()

print(x)

#same method with lowercase()

txt = "Hello, And Welcome To My World!"

x = txt.lower()

print(x)

#looping through index of the string
#range() function will return the range of values ... it works for both string and numbers ... (string means the index value it takes)

t = 'World'

for i in range(len(t)):
    print(t[i])

print('------------------------------------------------------------------------------')
#string iteration without the index value 

s = 'Buzz'

for char in s:
    print(char)

#string center() method example.
#The center() method will center align the string, using a specified character (space is default) as the fill character.

txt = "banana"

x = txt.center(20)

print(x)

print('-------------------------------------------------------------------------------')
#example2
#string.center(length, character)-- length is required space and charecter is to fill the empty space.

txt = "banana"

x = txt.center(20, "x")

print(x)

print('------------------------------------------------------------------------------')

#count() this wil returns number of charecters and words as for requirement.
#example
txt = "I love apples, apple are my favorite fruit"

x = txt.count("apple")
y=txt.count()
print(y)

print(x)
```
